```python
import cv2
import numpy as np
import matplotlib.pyplot as plt 
```

```python
img = cv2.imread(r'C:\Users\Suyash Tambe\Desktop\FIPL VS CODE\images.png')
img
```

```python
ret, thresh = cv2.threshold(img,180,255,2)
```

```python
thresh
```

```python
kernel = np.ones((5,5),np.uint8)
erosion = cv2.erode(thresh,kernel,iterations = 1)
```

```python
kernel
```

```python
fig = plt.figure(dpi=300)
plt.subplot(131)
plt.imshow(img,'gray')
plt.axis('off')
plt.title('Original Image')

plt.subplot(132)
plt.imshow(thresh,'gray')
plt.axis('off')
```

```python
fig=plt.figure(dpi=300)
plt.subplot(131)
plt.imshow(img,'gray')
plt.axis('off')
plt.title('Original image')

plt.subplot(132)
plt.imshow(thresh,'gray')
plt.axis('off')
plt.title('Threshold image')

plt.subplot(133)
plt.imshow(erosion,'gray')
plt.axis('off')
plt.title('Eroded Image')
```

```python
dilation = cv2.dilate(erosion,kernel,iterations=1)
```

```python
fig = plt.figure(dpi=300)
plt.subplot(221)
plt.imshow(img,'gray')
plt.axis('off')
plt.title('Original Image')

plt.subplot(222)
plt.imshow(thresh,'gray')
plt.axis('off')
plt.title('Binary image')

plt.subplot(223)
plt.imshow(erosion,'gray')
plt.axis('off')
plt.title('Eroded Image')

plt.subplot(224)
plt.imshow(dilation,'gray')
plt.axis('off')
plt.title('Dilated Image')
```

```python
opening = cv2.morphologyEx(thresh, cv2.MORPH_OPEN,kernel, iterations=1)
```

```python
fig=plt.figure(dpi=300)
plt.subplot(231)
plt.imshow(img,'gray')
plt.axis('off')
plt.title('Original image')

plt.subplot(232)
plt.imshow(thresh,'gray')
plt.axis('off')
plt.title('Threshold image')

plt.subplot(233)
plt.imshow(erosion,'gray')
plt.axis('off')
plt.title('Eroded Image')

plt.subplot(234)
plt.imshow(dilation,'gray')
plt.axis('off')
plt.title('Dilation Image')

plt.subplot(235)
plt.imshow(opening,'gray')
plt.axis('off')
plt.title('opened Image')
```

```python
closing = cv2.morphologyEx(thresh, cv2.MORPH_CLOSE, kernel)
```

```python
fig = plt.figure(dpi=300)
plt.subplot(131)
plt.imshow(img, 'gray')
plt.axis('off') 
plt.title('Original image')

plt.subplot(132)
plt.imshow(opening,'gray')
plt.axis('off')
plt.title('opened Image')

plt.subplot(133)
plt.imshow(closing,'gray')
plt.axis('off')
plt.title('Closed Image')
```
