```python
img = cv2.imread(r'C:\Users\Suyash Tambe\Desktop\FIPL VS CODE\IMG20230219225159.jpg',cv2.IMREAD_GRAYSCALE)
```

```python
mean = 0
stddev = 500
noise = np.zeros(img.shape, np.uint8)
cv2.randn(noise, mean, stddev)


noisy_img = cv2.add(img, noise)


cv2.imwrite(r'C:\Users\Suyash Tambe\Desktop\FIPL VS CODE\IMG20230219225159.jpg', noisy_img)
```

```python
Kernel = np.ones((5,5),np.float32)/25
dst = cv2.filter2D(noisy_img,-1, Kernel)

```

```python
plt.subplot(121)
plt.imshow(noise,'gray')
plt.subplot(122)
plt.hist(noise.ravel(),256,[0,256])
plt.show()
```
